```python
## Case Study – Student Registration System

student_name = input("Enter student_name :")
student_age = int(input("Enter student_age :"))
course_name = input("Enter course_name :")
course_fee = float(input("Enter course_fee :"))

gst = course_fee * 18/100
total_fee = course_fee + gst
print("\n *------ Student_Details:------*")
print(f"student_name - {student_name}")
print(f"student_age - {student_age}")
print(f"course_name - {course_name}")
print(f"course_fee - {course_fee}")
print("\n *------ Fees_Details:------ *")
print(f"gst - {gst}")
print(f"total_fee - {total_fee}")
print("\n *------- Data_type ------*")
print(type(student_name))
print(type(student_age))
print(type(course_name))
print(type(course_fee))

## Output
```

```python
## Case Study – Library Fine Calculator

student_name = input("Enter student_name - ")
book_name = input("Enter Book_name - ")
days_late = int(input("Enter days-late - "))
fine_per_day = float(input("Enter fine_per_day - "))
total_fine = days_late * fine_per_day
print("\n-----------Library Fine Details -----------")
print(f"student_name - {student_name}")
print(f"book_name - {book_name}")
print(f"days_late - {days_late}")
print(f"fine_per_day - {fine_per_day}")
print(f"total_fine - {total_fine}")
print("\n-----------Datatype Details -----------")
print(type(student_name))
print(type(book_name))
print(type(days_late))
print(type(fine_per_day))
print(type(total_fine))

##Output
```
