```python
# while loop

i = 1
while i <= 20:
 print(i)
 i += 1
print(f"Loop End till number reach 20")

```

```python
#Print all even numbers

i = 2
while i <= 20:
  print(i)
  i % 2 == 0
  i = i + 2
```

```python
#Print all odd numbers

i = 3
while i <= 20:
  print(i)
  i % 2 == 1
  i = i + 3
```

```python
#Case Study – ATM PIN Verification System

Correct_Pin = 1234
while True:
  Pin = int(input("Enter the Pin :"))
  if Correct_Pin == Pin:
   print("Login Successful")
   break
  else:
   print("Incorrect PIN... Try Again")
```

```python
#Case Study – Password Verification System with Limited Attempts

password = "Rekaa"
attempts = 0
max_attempts = 3

while attempts <= max_attempts:
 enter_pwd = input(f"Enter the password :")
 if enter_pwd == password:
   print("Access Granted")
   break
 else:
  attempts += 1
  print("Increase attempt count")

  if attempts == max_attempts:
    print("Access Denied")
```
