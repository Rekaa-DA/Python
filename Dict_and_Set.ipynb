```python
#Case Study – Student Data Management System

student = {"Name" : ["Ravi","Raja","kumar","Roja"],
           "age" : [25,23,24,26],
           "course details" : ["sql","excel","python","c"]
           }
print(student)

 # append

student["Name"].append("sam")
student["age"].append("23")
student["course details"].append("c++")

print(f"Name is :{student['Name']}")
print(f"Age is :{student['age']}")
print(f"course details is :{student['course details']}")


print("\n----- First Student -----")
print(f"Name: {student['Name'][0]}")
print(f"Age: {student['age'][0]}")
print(f"Course:{student['course details'][0]}")

print("\n----- New Student Index -----")
print(f"Name: {student['Name'][3]}")
print(f"Age: {student['age'][3]}")
print(f"Course:{student['course details'][3]}")



```

```python
# Case Study – Student Course Registration System using Set

course = {"Python", "SQL", "Power BI", "Python", "Excel"}
print(f"course details :{course}")
course.add("java")
print(f"Added Course :{course}")
course.update(["c","cea"])
print(f"Updated Course :{course}")
course.remove("SQL")
print(f"Removed Course :{course}")
course.discard("Excel")
print(f"discard Course :{course}")
print(f"Total Courses", len(course))

courses = {"Python", "SQL", "Power BI", "Python", "Excel"}

# Display original set
print("Original Courses:", courses)

# add() - Add one course
courses.add("Tableau")
print("\nAfter add():", courses)

# update() - Add multiple courses
courses.update(["AI", "Machine Learning"])
print("\nAfter update():", courses)

# remove() - Remove course
courses.remove("SQL")
print("\nAfter remove():", courses)

# discard() - Remove safely
courses.discard("Java")
print("\nAfter discard():", courses)

# Display total courses
print("\nTotal Courses:", len(courses))

```

```python
#tuple operations

Events =  ("Diwali", "Pongal", "Christmas", "Eid", "New Year")
print(Events)
print(f"First Event :{Events[0]}")
print(f"Last Event : {Events[-1]}")
print(f"Total Number of Events: {len(Events)}")

```
