1.Case Study – Shopping Cart System
Instructions:

Create a list named cart with initial items: Milk, Bread, Eggs, Apple

Perform list operations:

• Add multiple items using extend()

• Insert item at index 2 using insert()

• Remove item Apple using remove()

• Remove last item using pop()

Use list methods:

• index() to find position of Eggs

• count() to count occurrences of Milk

Apply indexing:

• Display first item

• Display last item

Apply slicing:

• Display first 3 items from the list

Find total number of items using len()

Display all outputs using print()

```python
#create list

cart = ["Milk", "Bread", "Eggs","Apple"]
print("-----------List Details---------\n")
print(f"List of cart :{cart} ")


#Adding Multiple items

cart.extend(["curd","soap","bun"])
print(f"List of including Extend_items  :{cart}")

#insert item

cart.insert(2,"mango")
print(f"Inserted item:{cart}")

#remove item

cart.remove("Apple")
print(f"Removed item:{cart}")

#last item removal

cart.pop()
print(f"Last Item Remove :{cart}")

# find position

index= cart.index("Eggs")
print(f"Index of Eggs:{index}")


# curd Count

print(f"Count of curd: {cart.count('curd')}")

#first item index

First_item = cart[0]
print(f"First item in list is : {First_item}")


#last item index

last_item = cart[-1]
print(f"Last item in list is : {last_item}")

# display first 3 items

three_items = cart[:3]
print(f"Three item in list is : {three_items}")

# total items
total_items = len(cart)
print(f"Total Items in List : {total_items}")

```

2.Case Study – Student Marks Analysis
Instructions:

Create a list named marks by getting 5 subject marks from the user using map() and list()

Perform calculations:

• Find total marks using sum()

• Find average marks using

• Calculate over all percentage assuming each subject is out of 100

```python
#Creating list

Student_Marks = list(map(int,input("Enter Marks :").split(",")))
print(Student_Marks)

#total marks using sum()

total_marks = sum(Student_Marks)
print(f"Total_marks is :{total_marks}")

#average marks using

Average_Mark = total_marks / len(Student_Marks)
print(f"Average Mark is : {Average_Mark}")

#over all percentage

percentage = (total_marks / (len(Student_Marks) * 100)) * 100
print(f"Percentage    : {percentage}")
```
