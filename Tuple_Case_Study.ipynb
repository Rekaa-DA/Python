```python
#Case Study – Online Order Summary using Tuple

# Packing
order = ("I01", "Mobile", 1, 55000, "Delivered")

# Unpacking
order_id, product_name, quantity, price, status = order

print(order_id)
print(product_name)
print(quantity)
print(price)
print(status)

#Output
```

```python
Order = (order_id, product_name, quantity, price, status)
print(Order)
print(type(Order))
```

```python
#Case Study - Student Details using Tuple (Packing & Unpacking)

# Packing
student = ("Roja", 22, "Excel")

print("Packed Tuple:", student)

# Unpacking (extracting values from tuple into variables)
name, age, course = student

print("Name:", name)
print("Age:", age)
print("Course:", course)


#Output
```
