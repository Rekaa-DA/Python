```python
#Case Study – Text Analyzer System

Text = input("Enter the Text:").lower()

vowels = 0
consonants = 0
digits = 0
special = 0

for T in Text:
  if(T in "aeiou"):
   vowels+=1
  elif(T.isalpha()):
   consonants+=1
  elif(T.isdigit()):
   digits+=1
  else:
   special+=1

print("Vowels:", vowels)
print("Consonants:", consonants)
print("Digits:", digits)
print("Special Characters:", special)
```

```python
# Case Study – Restricted Word Detection System

normal = ["yes","no","spam","evil","okay"]
restricted = ["evil"]

for word in normal:
  if(word==restricted):
    print("Restricted word detected")
    break
  print(word)
```

```python
# Case Study –Count of Restricted Word

normal = ["yes","no","spam","evil","okay"]
restricted = ["evil","no"]
new_words = []

count = 0
for w in normal:
  if(w in restricted):
   count += 1
  else:
    new_words.append(w)
print(f"List of new words : {new_words}")
print(count)
```

Create two lists: students marks

Use range() to: Access and print student names and marks using index positions

Use enumerate() to: Display student index number and student name

Use zip() to: Combine student names and marks together

Display all outputs using for loops

Use proper headings with print() statements

Print the final result in a clean format

```python
# Case Study – Student Performance Management System

students = ["Ravi","Raja","Roja","Sweetha","kumar"]
marks = [80,85,80,65,90]

print("\n--- Using range() ---")
for i in range(len(students)):
  print(f" Student name {students[i]} and mark is {marks[i]}")
print("\n--- Using enumerate() ---")
for i,s in enumerate(students):
  print(f"Index number id {i} for student {s}" )
print("\n--- Using zip() ---")
for s,m in zip(students,marks):
  print(f"Student name is {s} and mark {m}")
```
