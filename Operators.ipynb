```python
#Case Study: Bank Loan Eligibility System

name = input("Enter name:")
age = int(input("Enter Age:"))
income = int(input("Enter income:"))
credit_score = int(input("credit_score:"))

existing_loan = (input("Enter existing_loan Status:"))

if age >= 21 and income >=30000:
  if credit_score >= 700 and not existing_loan:
    print("Approved")
  elif credit_score >= 600 and income >= 50000:
    print("conditionally Approved")
  else:
    print("Rejected")

```
